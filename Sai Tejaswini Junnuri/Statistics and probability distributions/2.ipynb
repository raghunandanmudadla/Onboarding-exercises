Binomial discrete distribution: It is a probability distribution that summarizes the likelihood of a certain numer of succesess in a series of independent experimrnts under the assumptions that there is only one outcome for each trial, that each trial has the same probability of success, and that each trial is mutually exclusive, or independent of each other.
Ex: The probability of k heads in n independent toss throws(n>=k).

```python
import numpy as np
import matplotlib.pyplot as plt 
def f(n):
    t = 1
    for i in range(1,n+1):
        t = t*i
    return t
n= 20
p = 0.5
pm  =[]
x = []
for k in range(1,21):
    pm.append(round((f(n)/(f(n-k)*f(k)))*(p**k)*((1-p)**(n-k)),4))
for i in range(1,21):
    x.append(i)
plt.plot(x,pm,'o')
```

Poisson Discrete Distribution: It is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant rate and independently of the time since the last event.
Ex: The number of patients arriving in an emergency room between 10 and 11 pm.

```python
import numpy as np
import matplotlib.pyplot as plt 
def f(n):
    t = 1
    for i in range(1,n+1):
        t = t*i
    return t
average = 5
n  =20
a=[]
pm = []
for i in range(1,21):
    a.append(i)
for i in range(1,21):
    pm.append(((np.exp(-1*average))*(average**i))/f(i))
plt.plot(a,pm,'o')
```

Gamma Distribution: It is a continuous probability distribution that arises naturally in processes for which the waiting times between events are relevant.These distributions are useful in real-life where something has a natural minimum of 0.

```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
a = 1.99
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
x = np.linspace(gamma.ppf(0.01, a),gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a))
```

Normal Distribution: Also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.It is motivated by the Central Limit Theorem. This theory states that averages calculated from independent, identically distributed random variables have approximately normal distributions, regardless of the type of distribution from which the variables are sampled.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x))
```

Exponential Distribution: Exponential distribution or negative exponential distribution represents a probability distribution to describe the time between events in a Poisson process. In Poisson process events occur continuously and independently at a constant average rate. Exponential distribution is a particular case of the gamma distribution.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = expon.stats(moments='mvsk')
x = np.linspace(expon.ppf(0.01),expon.ppf(0.99), 100)
ax.plot(x, expon.pdf(x))
```

Chi Square Distribution: The chi-squared distribution (chi-square or X2 - distribution) with degrees of freedom, k is the distribution of a sum of the squares of k independent standard normal random variables.The chi-squared distribution is a special case of the gamma distribution.

```python
import numpy as np
from scipy.stats import chi
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
df=2
mean, var, skew, kurt = chi.stats(df,moments='mvsk')
x = np.linspace(chi.ppf(0.01,df),chi.ppf(0.99,df), 100)
ax.plot(x, chi.pdf(x,df))
```

Bayes Theorem: Let A1, A2, ... , An be a set of mutually exclusive events that together form the sample space S. Let B be any event from the same sample space, such that P(B) > 0. Then,
P( Ak | B ) =   P( Ak ) P( B | Ak ) /P( A1 ) P( B | A1 ) + P( A2 ) P( B | A2 ) + . . . + P( An ) P( B | An ) 
It is a theorem describing how the conditional probability of each of a set of possible causes for a given observed outcome can be computed from knowledge of the probability of each cause and the conditional probability of the outcome of each cause.

Confidence interval: It is the range in which you expect a certain percentage of output for the give data falls.

```python
from scipy.stats import sem, t
from scipy import mean
import numpy as np
confidence = 0.95
data = [1, 2, 3, 4, 5]

n = len(data)
m = mean(data)
std_err = (np.std(data)/2)
h = std_err * t.ppf((1 + confidence) / 2, n - 1)
print([m-h,m+h])
```

Significance Testing: Significance tests give us a formal process for using sample data to evaluate the likelihood of some claim about a population value. The help us in determining whether to reject the null hypothesis or accept the alternate hypothesis.

T test: A t-test is a type of inferential statistic used to determine if there is a significant difference between the means of two groups, which may be related in certain features. This test calculates the P-value of a sample vs a normal population or vs another sample. The result, P-value, tells you how likely these samples comes from the exact same distribution.

Kolmogorov-smirnov test: The KolmogorovSmirnov statistic quantifies a distance between the empirical distribution function of the sample and the cumulative distribution function of the reference distribution, or between the empirical distribution functions of two samples. Since the sample mean and standard deviation can be highly similar for different distributions, the Students T-Test gives a very high p-value in those cases too, whereas KS Test can detect the variance.

ANOVA: Analysis of variance (ANOVA) is a collection of statistical models and their associated estimation procedures (such as the "variation" among and between groups) used to analyze the differences among group means in a sample. In other words, they help you to figure out if you need to reject the null hypothesis or accept the alternate hypothesis. Basically, youre testing groups to see if theres a difference between them.

Chi Square Test: A chi-square goodness of fit test determines if a sample data matches a population. For more details on this type, see: Goodness of Fit Test.
A chi-square test for independence compares two variables in a contingency table to see if they are related. In a more general sense, it tests to see whether distributions of categorical variables differ from each another. 
A very small chi square test statistic means that your observed data fits your expected data extremely well. In other words, there is a relationship. 
A very large chi square test statistic means that the data does not fit very well. In other words, there isnt a relationship.

Correlation: Correlation is a statistical technique that can show whether and how strongly pairs of variables are related. In the broadest sense correlation is any statistical association, though it commonly refers to the degree to which a pair of variables are linearly related.

Covariance: Covariance is a measure of the joint variability of two random variables. If the greater values of one variable mainly correspond with the greater values of the other variable, and the same holds for the lesser values, the covariance is positive. Covariance provides a measure of the strength of the correlation between two or more sets of random variates.
