Mean: Mean is the central value of a discrete set of numbers i.e., the sum of values divided by the number of values

```python
import numpy as np
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = np.mean(x)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(y,'o', color  = 'blue')
```

Median: Median is the middle value in a distribution in the sense that half of the values are below and half are above median.

```python
import numpy as np
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = np.median(x)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(y,'o', color  = 'blue')
```

Mode: Mode is the value that appears most often in a data set, in a discrete probability distribution, it is where the probability mass function has maximum value.

```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
counter = {}
for i in x:
    counter.setdefault(i, 0)
    counter[i] += 1
mx = max(counter,key=counter.get)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(mx,'o', color  = 'blue')   
```

Range: Range is the sixe of the smallest interval that contains the data, henceforth shows statistical dispersion.

```python
import numpy as np
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = (min(x),max(x))
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(y[0],'o', color  = 'blue')
plt.plot(y[1],'o', color  = 'blue')
```

Standard deviation: Standard deviation is the measure of dispersion of a set of data from its mean. It measures the absolute variability of a distribution.

```python
import numpy as np
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = np.std(x)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(y,'o', color  = 'blue')
```

Variance: It measures how far data is spread out from their average value. It is the square of standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = np.var(x)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(y,'o', color  = 'blue')
```

Population mean is the average of the entire data. Sample mean is the average  of the randomly selected sample from the entire data.


```python
import numpy as np
import random
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
y = random.sample(x,5)
z = np.mean(x)
k = np.mean(y)
for i in x:
    plt.plot(i,'o',color= "green")
plt.plot(z,'o', color  = 'blue',label = 'population mean')
plt.plot(k,'o', color  = 'red',label = 'sample mean')
```

Chebyshevs theorem: Chebyshevs theorem states that no more than 1/k of the distribution's values can be more than k standard deviations away from the mean for any given k which is greater than 1

```python
import numpy as np
import random
import matplotlib.pyplot as plt 
x=[2,3,3,4,5,6,10,14]
k =4
m = np.mean(x)
std = np.std(x)
num = 0
r = [m-(k*std),m+(k*std)]
for i in x:
    if (i>=r[0])and(i<=r[1]):
        num = num+1
if (len(x)-num)<=(1/(k**2))*(num):
    print(True)
else:
    print(False)
```

Central limit theorem: The Central Limit Theorem is a statistical theory that states the distribution of sample means approximates a normal distribution as the sample size gets larger (assuming that all samples are identical in size), regardless of population distribution shape.
